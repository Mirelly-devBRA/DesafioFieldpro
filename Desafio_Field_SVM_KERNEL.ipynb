```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Carregando os dados do arquivo Sensor_FieldPRO.csv
sensor_data = pd.read_csv('Sensor_FieldPRO.csv')

# Carregando os dados do arquivo Estacao_Convencional.csv
estacao_data = pd.read_csv('Estacao_Convencional.csv')

# Combinando os dataframes
data_combined = pd.concat([sensor_data, estacao_data], axis=1)

# Selecionando as variáveis relevantes para a modelagem
relevant_features = ['atm_pressure_main', 'num_of_resets', 'piezo_charge', 'piezo_temperature']
target_variable = 'chuva'

data_selected = data_combined[relevant_features + [target_variable]]

# Tratando valores ausentes preenchendo com a média
data_selected.fillna(data_selected.mean(), inplace=True)

# Padronizando as variáveis numéricas
scaler = StandardScaler()
data_selected[relevant_features] = scaler.fit_transform(data_selected[relevant_features])

# Separando os dados em conjunto de treinamento e teste
X = data_selected[relevant_features]
y = data_selected[target_variable]

# Definindo a proporção de dados para o conjunto de teste (por exemplo, 20%)
test_size = 0.2

# Dividindo os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

# Treinando o modelo SVM para regressão
svm_model = SVR(kernel='linear')
svm_model.fit(X_train, y_train)

# Fazendo previsões no conjunto de teste
y_pred = svm_model.predict(X_test)

# Avaliando o desempenho do modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse}")
print(f"R²: {r2}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Carregando os dados do arquivo Sensor_FieldPRO.csv
sensor_data = pd.read_csv('Sensor_FieldPRO.csv')

# Carregando os dados do arquivo Estacao_Convencional.csv
estacao_data = pd.read_csv('Estacao_Convencional.csv')

# Combinando os dataframes
data_combined = pd.concat([sensor_data, estacao_data], axis=1)

# Selecionando as variáveis relevantes para a modelagem
relevant_features = ['atm_pressure_main', 'num_of_resets', 'piezo_charge', 'piezo_temperature']
target_variable = 'chuva'

data_selected = data_combined[relevant_features + [target_variable]]

# Tratando valores ausentes preenchendo com a média
data_selected.fillna(data_selected.mean(), inplace=True)

# Padronizando as variáveis numéricas
scaler = StandardScaler()
data_selected[relevant_features] = scaler.fit_transform(data_selected[relevant_features])

# Separando os dados em conjunto de treinamento e teste
X = data_selected[relevant_features]
y = data_selected[target_variable]

# Definindo a proporção de dados para o conjunto de teste (por exemplo, 20%)
test_size = 0.2

# Dividindo os dados em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

# Definindo os hiperparâmetros para calibrar o modelo
param_grid = {
    'kernel': ['linear', 'rbf', 'poly'],  # Experimente diferentes kernels
    'C': [0.1, 1, 10],  # Experimente diferentes valores de C
    'epsilon': [0.01, 0.1, 1]  # Experimente diferentes valores de epsilon (apenas para kernel RBF)
}

# Criando o modelo SVM para regressão
svm_model = SVR()

# Realizando a busca exaustiva pelos melhores hiperparâmetros usando validação cruzada
grid_search = GridSearchCV(svm_model, param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X_train, y_train)

# Obtendo os melhores hiperparâmetros
best_params = grid_search.best_params_
print("Melhores Hiperparâmetros:")
print(best_params)

# Obtendo o modelo com os melhores hiperparâmetros
best_svm_model = grid_search.best_estimator_

# Fazendo previsões no conjunto de teste usando o modelo calibrado
y_pred = best_svm_model.predict(X_test)

# Avaliando o desempenho do modelo calibrado
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse}")
print(f"R²: {r2}")
```

```python
# Plotando os resultados
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--')
plt.xlabel('Valores Reais')
plt.ylabel('Previsões')
plt.title('Comparação entre Valores Reais e Previsões (SVM)calibrado')
plt.show()
```
