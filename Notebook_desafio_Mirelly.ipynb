```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Carregar os dados dos arquivos CSV (caso ainda não tenha feito isso)
sensor_data = pd.read_csv("Sensor_FieldPRO.csv")
estacao_data = pd.read_csv("Estacao_Convencional.csv")

# Verificar se há valores faltantes nos DataFrames
print("Valores faltantes no DataFrame Sensor_FieldPRO:")
print(sensor_data.isnull().sum())

print("\nValores faltantes no DataFrame Estacao_Convencional:")
print(estacao_data.isnull().sum())

# Preencher valores faltantes com a média ou mediana (dependendo da distribuição dos dados)
sensor_data.fillna(sensor_data.mean(), inplace=True)
estacao_data.fillna(estacao_data.mean(), inplace=True)

# Normalizar ou padronizar as variáveis numéricas (caso necessário)
# Normalização coloca os dados em uma escala entre 0 e 1
# Padronização coloca os dados com média 0 e desvio padrão 1

# Variáveis numéricas para normalizar ou padronizar
cols_to_normalize = ['piezo_charge', 'piezo_temperature', 'num_of_resets', 'air_temperature_100', 'air_humidity_100', 'atm_pressure_main']

# Criar um StandardScaler para padronizar as variáveis
scaler = StandardScaler()

# Aplicar a padronização nos dados do sensor
sensor_data[cols_to_normalize] = scaler.fit_transform(sensor_data[cols_to_normalize])

# Aplicar a padronização nos dados da estação convencional
estacao_data[cols_to_normalize] = scaler.transform(estacao_data[cols_to_normalize])

# Verificar os DataFrames após o pré-processamento
print("\nDataFrame Sensor_FieldPRO após pré-processamento:")
print(sensor_data.head())

print("\nDataFrame Estacao_Convencional após pré-processamento:")
print(estacao_data.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score

# Carregar os dados das estações meteorológicas (neste exemplo, usaremos dois arquivos CSV, mas você pode adicionar mais conforme necessário)
estacao_data1 = pd.read_csv("Estacao_Convencional1.csv")
estacao_data2 = pd.read_csv("Estacao_Convencional2.csv")

# Combinar os dados das diferentes estações em um único DataFrame
all_estacoes_data = pd.concat([estacao_data1, estacao_data2], ignore_index=True)

# Realizar o pré-processamento dos dados (se necessário)

# Separar as variáveis preditoras (X) e a variável alvo (y) no DataFrame combinado
X_all_estacoes = all_estacoes_data.drop(columns=['piezo_charge'])  # Variáveis preditoras das estações
y_all_estacoes = all_estacoes_data['piezo_charge']  # Variável alvo (carga do acumulador) das estações

# Ajustar o modelo de Regressão Linear usando Lasso Regression para o conjunto de todas as estações
lasso_model_all_estacoes = Lasso(alpha=0.1)  # O valor de alpha controla o nível de penalização L1
lasso_model_all_estacoes.fit(X_all_estacoes, y_all_estacoes)

# Realizar a validação cruzada com o modelo ajustado usando a função cross_val_score
# Neste exemplo, usaremos 5 folds de validação cruzada (cv=5), mas você pode ajustar o número conforme necessário
cv_scores = cross_val_score(lasso_model_all_estacoes, X_all_estacoes, y_all_estacoes, cv=5, scoring='neg_mean_squared_error')

# Calcular o MSE médio e o desvio padrão dos resultados da validação cruzada
mse_mean = -cv_scores.mean()
mse_std = cv_scores.std()

print("MSE médio da validação cruzada:", mse_mean)
print("Desvio padrão do MSE da validação cruzada:", mse_std)

# Outra métrica para avaliar o desempenho é o R²
r2_scores = cross_val_score(lasso_model_all_estacoes, X_all_estacoes, y_all_estacoes, cv=5, scoring='r2')
r2_mean = r2_scores.mean()

print("R² médio da validação cruzada:", r2_mean)
```
